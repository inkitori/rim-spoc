```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import math

# 1. Your tensor data
# (I'll copy it here for completeness, but in your script, you'd just use your existing tensor)
# Create a PyTorch tensor.
# If your tensor is already a PyTorch tensor on CUDA, use it directly.
# If it's a list of lists like above, convert it:
attention_tensor = torch.tensor(attention_data, device='cuda:0') # Assuming it was on cuda as per your prompt

# 2. Ensure tensor is on CPU and is a NumPy array for Matplotlib
if attention_tensor.is_cuda:
    attention_numpy = attention_tensor.cpu().numpy()
else:
    attention_numpy = attention_tensor.numpy()

# 3. Get dimensions
num_heads, target_seq_len, source_seq_len = attention_numpy.shape

# 4. Determine grid size for subplots
# We want a somewhat square grid for the heads
ncols = math.ceil(math.sqrt(num_heads))
nrows = math.ceil(num_heads / ncols)

fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * 4, nrows * 3.5)) # Adjust figsize as needed
axes = axes.flatten() # Flatten to make it easier to iterate if nrows > 1 or ncols > 1

# Optional: If you have token labels for source and target sequences
# source_tokens = [f"Src {i+1}" for i in range(source_seq_len)]
# target_tokens = [f"Tgt {i+1}" for i in range(target_seq_len)]
# If not, we'll just use indices.
source_labels = [str(i) for i in range(source_seq_len)]
target_labels = [str(i) for i in range(target_seq_len)]


for i in range(num_heads):
    ax = axes[i]
    head_attention = attention_numpy[i, :, :]

    # Display the heatmap
    # You can experiment with different colormaps: 'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'hot', 'gray'
    im = ax.imshow(head_attention, cmap='viridis', aspect='auto') # 'aspect="auto"' can be useful if seq lengths differ

    ax.set_title(f"Head {i + 1}")

    # Set ticks and labels
    ax.set_xticks(np.arange(source_seq_len))
    ax.set_yticks(np.arange(target_seq_len))

    ax.set_xticklabels(source_labels)
    ax.set_yticklabels(target_labels)

    ax.set_xlabel("Source Sequence Position")
    ax.set_ylabel("Target Sequence Position")

    # Rotate x-axis labels if they overlap
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    # Add a colorbar to each subplot
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

# Hide any unused subplots if num_heads is not a perfect square/rectangle
for j in range(num_heads, nrows * ncols):
    fig.delaxes(axes[j]) # or axes[j].axis('off')

plt.tight_layout() # Adjusts subplot params for a tight layout
plt.suptitle("Attention Heatmaps per Head", fontsize=16, y=1.02) # y to position above tight_layout
plt.show()
```
